#### Convert to various types form Qiime2

```python
#Table
table = q2.Artifact.import_data('FeatureTable[Frequency]', os.getcwd() + '/qiita_downloads/qiita14458_NormalGerdEsoph/169125_none_WOL.biom')

#WOL Taxonomy
taxdf = pd.read_csv('/Users/cguccion/Dropbox/Storage/HelpfulLabDocs/taxonomy_trees/WOL/lineages.txt', sep='\t', index_col=0, header=None)
taxdf.columns = ['Taxon']
taxdf.index.name = 'Feature ID'
taxonomy = q2.Artifact.import_data('FeatureData[Taxonomy]', taxdf)

collapsed_table = taxa.methods.collapse(table=table, taxonomy=taxonomy, level=7).collapsed_table

#display(collapsed_table.view(pd.DataFrame).T)
Artifact.export_data(collapsed_table, 'species_WOL')
```
