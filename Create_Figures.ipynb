```python
def plot_alpha(df, key_col):
    
    order = ['normal_genome_WOL_scrubbed', 'GERD_genome_WOL_scrubbed',
       'BE_NP_genome_WOL_scrubbed', 'BE_Prog_T1_genome_WOL_scrubbed',
       'BE_Prog_T2_genome_WOL_scrubbed',
       'BE_exact_Ross_paired_genome_WOL_scrubbed',
       'EAC_Ross_paired_genome_WOL_scrubbed', 'EAC_ICGC_genome_WOL_scrubbed']
    
    plt.figure(figsize=(10, 8)) 
    
    sns.set_theme(style="white")
    # Create custom axis
    ax = sns.boxplot(y=key_col, x='study_label', data=df, order=order)
    
    # Strip plot
    sns.stripplot(y=key_col, x='study_label', data=df, color='black', size=4, jitter=True, order=order, ax=ax)

    # Manually set tick labels
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    
    plt.tight_layout()
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']
    plt.rcParams['svg.fonttype'] = 'none'  # render SVG text as text, not curves
    #plt.xticks(rotation=45, ha='right')  
    #plt.xticks(rotation=90)  # Rotate x-axis labels
    plt.savefig('outputs/alpha_plots/alphaMod_Combo-Progression_WOL_scrubbed_' + key_col + '.svg')
    plt.show()
```
